# Exploratory Data Analysis

## Overall Customer Churn

```python
churn_counts = client_df['churn'].value_counts()
churn_percentages = churn_counts / len(client_df) * 100

plt.figure(figsize=(8, 6))
bars = plt.bar(['No Churn', 'Churn'], churn_counts, color=['skyblue', 'salmon'])

for bar, percentage in zip(bars, churn_percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, f'{percentage:.1f}%', ha='center', va='center', color='white', fontsize=12)


plt.xlabel('Churn Status')
plt.ylabel('Number of Clients')
plt.title('Overall Churn Distribution')
plt.show()
```
